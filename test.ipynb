```python
import sqlite3
import pandas as pd

dbConn = sqlite3.connect('nfl_data.db')
dbCursor = dbConn.cursor()
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

query = """
SELECT *
FROM week1
LIMIT 5;
"""

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
SELECT x, y, event, frameID, time
FROM week1
WHERE displayName = "Julio Jones" AND playID = "75"
"""

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
SELECT DISTINCT week1.time, playID, week1.x, week1.y, frameID, week1.event, pbp.series, week1.displayName, roster2018.full_name, playDirection
FROM pbp
JOIN week1 ON week1.gameID = pbp.old_game_id
JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
WHERE gameID = 2018090600 AND complete_pass = 1 AND week1.displayName = roster2018.full_name
ORDER BY pbp.series DESC
LIMIT 10;
"""

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT play_id, full_name, qtr, time, posteam, receiver_jersey_number, drive, series
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE old_game_id = 2018090900 AND posteam != "None" AND complete_pass = 1
        ORDER BY qtr ASC, time DESC;
        """

query2 = """
        SELECT play_id, full_name, receiver_jersey_number, qtr, time, posteam, series
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE old_game_id = ? AND posteam != "None" AND complete_pass = 1
        ORDER BY qtr ASC, time DESC;
        """

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT play_id, full_name, qtr, time, posteam, drive, desc, yards_gained, down, play_type, yrdln, complete_pass
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id OR pbp.rusher_player_id = roster2018.gsis_id
        WHERE old_game_id = 2018090600 AND posteam != "None" AND drive = 1
        ORDER BY qtr ASC, time DESC;
        """

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
SELECT DISTINCT event
FROM week1;
"""

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT play_id, full_name, desc, pbp.week, receiving_yards, yards_after_catch, air_yards
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE full_name = "Taylor Gabriel" AND posteam != "None" AND complete_pass = 1 AND pbp.week <= 17
        ORDER BY pbp.week ASC;
        """
dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT full_name, roster2018.jersey_number, team, position, sum(receiving_yards), sum(yards_after_catch), sum(complete_pass), count(complete_pass), sum(touchdown)
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE full_name = "Julio Jones" AND pbp.week <= 17;
        """
dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT full_name, receiver_player_name, desc, route, play_id
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        JOIN week1 ON pbp.play_id = week1.playId
        WHERE full_name = "Julio Jones" AND complete_pass = 1 AND receiver_player_name = "J.Jones"
        GROUP BY play_id
        LIMIT 5;
        """

dbCursor.execute(query)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```

```python
query = """
        SELECT receiver_player_name,
                sum(receiving_yards), RANK () OVER (ORDER BY sum(receiving_yards) DESC) AS recYardsRank, 
                sum(complete_pass), RANK () OVER (ORDER BY sum(complete_pass) DESC) AS recRank,
                sum(touchdown), RANK () OVER (ORDER BY sum(touchdown) DESC) AS tdRank
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE pbp.week <= 17
        GROUP BY receiver_player_id;
        """

query2 = """
        SELECT receiver_player_name,
                sum(complete_pass),
                sum(receiving_yards),
                sum(touchdown), 
                roster2018.position,
                RANK () OVER (ORDER BY sum(complete_pass) DESC) AS recRankAll,
                RANK () OVER (PARTITION BY roster2018.position ORDER BY sum(complete_pass) DESC) AS recRankPos,
                RANK () OVER (ORDER BY sum(receiving_yards) DESC) AS recYardRankAll,
                RANK () OVER (PARTITION BY roster2018.position ORDER BY sum(receiving_yards) DESC) AS recYardRankPos,
                RANK () OVER (ORDER BY sum(touchdown) DESC) AS tdRankAll,
                RANK () OVER (PARTITION BY roster2018.position ORDER BY sum(touchdown) DESC) AS tdRankPos
        FROM pbp
        JOIN roster2018 ON pbp.receiver_player_id = roster2018.gsis_id
        WHERE pbp.week <= 17
        GROUP BY receiver_player_id;
        """


dbCursor.execute(query2)
df = pd.DataFrame(dbCursor.fetchall())
df.head(len(df))
```
